# Problem Statement:

Small wineries often face the challenge of not being able to afford top wine testers to evaluate their products. Without such expertise, they may struggle to get an accurate assessment of the quality of their wines, which could affect their competitiveness in the market.For this reason I've created this model which can get a review from a local wine tester or a wine expert and might be able to get the a proper evaluation.

## Encoding Data

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
newdf_copy=new_df.copy()
final_df=new_df.copy()
enc.fit(new_df[['country','province','variety']])
new_df[['country','province','variety']]=enc.transform(new_df[['country','province','variety']])
```

```python
%%time
from tqdm import tqdm
from nltk.sentiment import SentimentIntensityAnalyzer
# nltk.download('vader_lexicon')
sia= SentimentIntensityAnalyzer()
data={}
for i,row in tqdm(enumerate(new_df.iterrows()),total=len(new_df)):
    id=row[0]
    description=row[1]['description']
    data[id]=sia.polarity_scores(description)
```

```python
new_data=pd.DataFrame(data).T
corr_df=pd.merge(new_df,new_data,left_index=True,right_index=True)
corr_df.tail()
```

```python
final_df.shape
```

```python
from sklearn.utils import resample
df_majority=final_df[(final_df['points']>=85)&(final_df['points']<=91)]
df_minority =final_df[(final_df['points']<85)|(final_df['points']>91)]
df_minority_res=resample(df_minority, 
                                 replace=True,    # sample without replacement
                                 n_samples=len(df_majority),     # to match minority class
                                 random_state=42)

final_df = pd.concat([df_minority_res, df_majority],axis='index').reset_index(drop=True)
X=final_df[['country','price','province','variety','description']]
y=final_df['points']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50,random_state=42)

final_df.shape,len(df_majority),len(df_minority)
```

# Model Training

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn import set_config #visulazie pipeline
from tqdm import tqdm
from nltk.sentiment import SentimentIntensityAnalyzer
from xgboost import XGBRegressor
from sklearn.metrics import r2_score


set_config(display="diagram")
set_config(transform_output="pandas")

# Define the function to add a new column
def add_column(X):
    data={}
    for i,row in tqdm(enumerate(X.iterrows()),total=len(X)):
        id=row[0]
        description=row[1]['remainder__description']
        data[id]=sia.polarity_scores(description)
    new_data=pd.DataFrame(data).T
    X.drop(['remainder__description'],axis=1,inplace=True)
    X=pd.merge(X,new_data,left_index=True,right_index=True)    
    return X
    

encoding = ColumnTransformer([
    ('Encoding',OrdinalEncoder(categories=enc.categories_),[0,2,3]),
],remainder='passthrough')
encoding.set_output(transform="pandas")
# Create a pipeline with the function transformer
pipe = Pipeline([
    ('Encoder',encoding),
    ('Add_description_scores', FunctionTransformer(add_column)),
    ('Regressor',XGBRegressor(max_depth=11,n_estimators=500, colsample_bytree=0.4,eta=0.1))
])

daa=pipe.fit(X_train,y_train)
pipe
```

```python
# pipe.named_steps['Encoder'].named_transformers_['Encoding'].categories_
```

```python
y_pred=pipe.predict(X_test)
y_pred
```

```python
score=r2_score(y_test,y_pred)
score
```

```python
cols=pipe.named_steps['Regressor'].get_booster().feature_names
cols
```
